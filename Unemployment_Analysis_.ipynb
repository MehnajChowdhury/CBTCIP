**Problem Statement:**

Unemployement is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force.We have seen a sharp increase in the unemployment rate during Covid-19,so analyzing the unemployment rate can be a good data science project.

```python
#Importing the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Loading the dataset of Unemployment_Rate_upto_11_2020 as csv

data=pd.read_csv('/Unemployment_Rate_upto_11_2020.csv')
data
```

```python
#Gathering the information about the dataset using info() method

data.info()
```

```python
#Getting the necessary descriptions about the dataset using describe() method

data.describe()
```

```python
#Checking for null or missing values in the dataset

data.isnull().sum()
```

```python
#Renaming the columns

data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
             'Estimated Employed','Estimeated Labour Participation Rate',
             'Region','Longitude','Latitude']
```

```python
#Analysing the top rows of the dataset using head() method

data.head()
```

**Checking correlation between the features of  the dataset**

```python
#Plotting the correlation heatmap of the dataset

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,10))
sns.heatmap(data.corr())
plt.show()
```

```python
#Plotting histplot to veiw estimated number of employee in different regions of India

data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
             'Estimated Employed','Estimeated Labour Participation Rate',
             'Region','Longitude','Latitude']
plt.title('Unemployement in India')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

```python
#Plotting histplot to veiw the unemployment rate in different regions of India

plt.figure(figsize=(12,10))
plt.title("Unemployement in India")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=data)
plt.show()
```

```python
#Creating a dashboard to analyze the unemployment rate in each state of India
#using sunburst plot

unemployment=data[['States','Region','Estimated Unemployment Rate']]
figure=px.sunburst(unemployment,path=["Region","States"],
                  values="Estimated Unemployment Rate",
                  width=700,height=700,color_continuous_scale="RdY1Gn",
                  title="Rate of Unemployment in India"
                  )
figure.show()
```
